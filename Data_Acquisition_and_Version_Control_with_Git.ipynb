#Data Acquisition & Exploration


## 1. Dataset Selection
I selected the **Iris dataset** from the UCI Machine Learning Repository. Collected by R.A. Fisher in 1936, it includes measurements of 150 iris flowers (sepal and petal dimensions) across three species.  
- **Collection Method**: Manual measurements by researchers.  
- **Reliability**: Published in a peer-reviewed journal, widely used in machine learning education.  
- **Limitations**: Small sample size (150 records) lacks real-world variability (e.g., environmental factors).  
- **Biases**: Data from one location (Quebec, Canada) may not represent global iris populations.

---
## 2. Data Import
The Iris dataset was imported via URL for reproducibility. Alternative methods (local upload or Google Drive) were tested but not used in this notebook.


```python
# Import pandas for data handling
import pandas as pd

# Load dataset into a DataFrame
column_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
df = pd.read_csv('iris.csv', header=None, names=column_names)

# Display first few rows to verify
df.head()
```

---
## 3. Data Exploration
The dataset has 150 rows and 5 columns: sepal_length, sepal_width, petal_length, petal_width, and species. All columns are numeric except species (object).

**Insight**: The scatter plot shows distinct clusters by speciesâ€”Setosa has smaller petals, while Virginica has larger ones. This suggests petal measurements are effective for species classification, but applying this model to other flower types could introduce bias if their measurements differ.

```python
# Import libraries for visualization
import matplotlib.pyplot as plt
import seaborn as sns  # Install if needed: !pip install seaborn

# Display dataset shape, columns, and data types
print("Dataset Shape:", df.shape)  # Rows and columns
print("Column Names:", df.columns.tolist())
print("Data Types:\n", df.dtypes)

# Summary statistics
print("\nSummary Statistics:\n", df.describe())

# Check for missing values
print("\nMissing Values:\n", df.isnull().sum())

# Visualization 1: Histogram of sepal length
plt.figure(figsize=(8, 4))
sns.histplot(df['sepal_length'], kde=True, color='teal')
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()

# Visualization 2: Scatter plot of petal length vs width by species
plt.figure(figsize=(8, 4))
sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=df, palette='deep')
plt.title('Petal Length vs Width by Species')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.show()
```

## 4. Ethical & Legal Reflection
The Iris dataset is public domain, allowing unrestricted use, modification, and distribution. It contains no personal data (only plant measurements), so privacy concerns are minimal, and GDPR does not apply as no EU personal data is involved. If similar datasets included location tags from EU sources, GDPR would require consent and data minimization. Potential biases include the datasetâ€™s origin (Quebec, Canada), which may not represent global iris variations, potentially skewing models for broader use. Additionally, equal species distribution assumes balanced classes, but real-world populations may differ, risking biased predictions. Ethically, users should avoid overgeneralizing results to unrepresented populations and ensure fair AI applications by validating models on diverse data.
